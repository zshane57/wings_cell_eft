```python
def show_img_grid(img1):
    image = Image.open(img1)
    image = np.array(image.convert("RGB"))

    plt.figure(figsize=(10, 10))
    plt.imshow(image)
    #show_anns(masks)
    plt.axis('on')
    plt.title(img1[47:-11])
    print(img1)
    plt.show()
```
